(3) CNN 모델 정의

CNN 모델은 여러 층으로 구성되며 주요 층은 다음과 같다.

Conv2D (Convolutional Layer): 이미지에서 특징을 추출

MaxPooling2D (Pooling Layer): 특징 맵을 다운샘플링하여 계산량을 줄이고 중요한 특징을 강조

Flatten: 다차원 배열을 1차원으로 변환

Dense (Fully Connected Layer): 분류 작업을 수행

Dropout: 과적합을 방지하기 위해 일부 뉴런을 무작위로 비활성화

(4) 모델 훈련:
훈련 데이터로 모델을 훈련하고, 검증 데이터로 모델을 평가한다.

(5) 모델 평가 및 시각화:
훈련 과정에서의 정확도와 손실을 시각화하여 모델의 성능을 평가합니다.

(6) 네트워크 구조

Conv2D (32 filters, 3x3 kernel, ReLU activation): 입력 이미지에서 32개의 특징 맵 추출

MaxPooling2D (2x2 pool size): 2x2 영역에서 최대값을 선택하여 다운샘플링

Conv2D (64 filters, 3x3 kernel, ReLU activation): 더 많은 특징 추출

MaxPooling2D (2x2 pool size): 다운샘플링 반복

Conv2D (128 filters, 3x3 kernel, ReLU activation): 고차원 특징 추출

MaxPooling2D (2x2 pool size): 마지막 다운샘플링 수행

Flatten: 3차원 배열을 1차원으로 변환하여 fully connected layer에 입력으로 제공

Dense (512 units, ReLU activation): 분류기를 구성하는 첫 번째 층

Dropout (0.5): 과적합을 방지하기 위해 50%의 뉴런을 무작위로 비활성화

Dense (1 unit, Sigmoid activation): 이진 분류 수행

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
!git clone https://github.com/sonyerim/DPR.git
```

```python
base_dir = '/content/DPR/images'
car_dir = os.path.join(base_dir, 'car')
bus_dir = os.path.join(base_dir, 'bus')
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    validation_split=0.2,
)
```

```python
train_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
```

```python
validation_generator = train_datagen.flow_from_directory(
    base_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    epochs=20
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(20)
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
